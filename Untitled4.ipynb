```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


sports_data = pd.read_csv('/content/Sports Details.csv')


entities = sports_data['Entity']
female_participants = sports_data['Female']
male_participants = sports_data['Male']

x = np.arange(len(entities))
width = 0.35

fig, ax = plt.subplots(figsize=(12, 6))
bars_female = ax.bar(x - width/2, female_participants, width, label='Female', color='lightblue')
bars_male = ax.bar(x + width/2, male_participants, width, label='Male', color='salmon')

ax.set_xlabel('Sports', fontsize=12)
ax.set_ylabel('Number of Participants', fontsize=12)
ax.set_title('Male vs Female Participation in Sports', fontsize=16)
ax.set_xticks(x)
ax.set_xticklabels(entities, rotation=45, ha='right', fontsize=10)
ax.legend()

plt.tight_layout()
plt.show()
```
